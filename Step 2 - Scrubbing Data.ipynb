```python
# Import
import pandas as pd
```

```python
df = pd.read_csv("IMDb_rating_data.csv",index_col = 0)
df.head()
```

```python
## Fill null\missing
df.isnull().sum()
```

```python
df['Gross'] = df['Gross'].str.replace('$', '', regex=True).replace('M', '', regex=True).astype(float)
df
```

```python
df['Gross'].fillna(format(df['Gross'].mean(), ".2f"),inplace = True)
df
```

```python
df.isnull().sum()
```

```python
## drop duplicates
df.drop_duplicates(subset='Name', keep='first', inplace=True)
df
```

```python
df.to_csv("IMDb_rating_data_without_nan.csv")
```
