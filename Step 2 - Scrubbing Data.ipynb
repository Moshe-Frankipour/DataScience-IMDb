## Step 2 - Scrubbing Data

In this step, we will Cleaning, formatting and filtering the data.

```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv("IMDb_rating_data.csv",index_col = 0)
df.head()
```

```python
## Fill null\missing
df.isnull().sum()
```

```python
df['Gross'] = df['Gross'].str.replace('$', '', regex=True).replace('M', '', regex=True).astype(float)
df
```

```python
df['Gross'].fillna(format(df['Gross'].mean(), ".2f"),inplace = True)
df
```

```python
df.isnull().sum()
```

```python
if (len(df[df.duplicated() == True]) > 0):
    print ("Droping Duplicates")
    df.drop_duplicates(inplace=True)
```

```python
print ("Drop ID from table")
df = df.drop('ID', axis=1)
df
```

```python
df.info()
```

Fixing objects to number

-For Votes:

```python
if (df['Votes'].dtype == object):
    df["Votes"] = [int(val.replace(',','')) for val in df["Votes"]]
df['Votes'].dtype
```

-For Gross:

```python
if (df['Gross'].dtype == object):
    df["Gross"] = [float(val) for val in df["Gross"]]
df['Gross'].dtype
```

-For MalesCount:

```python
if (df['malesCount'].dtype == object):
    df["malesCount"] = [int(val.replace(',','')) for val in df["malesCount"]]
df['malesCount'].dtype
```

-For FemalesCount:

```python
if (df['femalesCount'].dtype == object):
    df["femalesCount"] = [int(val.replace(',','')) for val in df["femalesCount"]]
df['femalesCount'].dtype
```

-For Year:

```python
if (df['Year'].dtype == object):
    df["Year"] = [int(re.sub("[^0-9]", "", val)) for val in df["Year"]]
df['Year'].dtype
```
